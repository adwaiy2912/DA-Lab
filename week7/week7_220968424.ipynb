```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("employment.csv", index_col=0)
df.head()
```

```python
df.isna().sum()
```

#### 2. Generate a boxplot to find the distribution of unemployment rate for every industry .

```python
plt.figure(figsize=(10,6))
sns.boxplot(df, orient="h")
plt.ylabel('Industry')
plt.xlabel('Unemployment Rate (%)')
plt.title('Distribution of Unemployment Rates by Industry')
plt.show()
```

#### 3. Using line chart Visualize the unemployment rate of workers by industry .

```python
plt.figure(figsize=(15,10))
pd.DatetimeIndex(df['date']).month_name()
sns.lineplot(df, x=df.resample('Y').mean().index.year)
plt.ylabel('Date')
plt.xlabel('Industry')
plt.legend()
```
